```python
# 📙 LEVEL 2 - TASK 1: Table Booking and Online Delivery Analysis

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\admin\\Downloads\\Dataset .csv")

# ✅ % of Restaurants with Table Booking
table_booking_percent = df['Has Table booking'].value_counts(normalize=True) * 100
print("Table Booking Availability (%):\n", table_booking_percent)

# ✅ % of Restaurants with Online Delivery
online_delivery_percent = df['Has Online delivery'].value_counts(normalize=True) * 100
print("\nOnline Delivery Availability (%):\n", online_delivery_percent)

# 📊 Compare Average Ratings: Table Booking vs Not
table_rating = df.groupby('Has Table booking')['Aggregate rating'].mean()
print("\nAvg Ratings (Table Booking):\n", table_rating)

# 📊 Online Delivery by Price Range
delivery_by_price = df[df['Has Online delivery'] == 'Yes'].groupby('Price range')['Restaurant ID'].count()
print("\nDelivery Availability Across Price Range:\n", delivery_by_price)

# 🎨 Visualizations
sns.barplot(x='Has Table booking', y='Aggregate rating', data=df)
plt.title("Table Booking vs Average Rating")
plt.show()

sns.countplot(x='Price range', hue='Has Online delivery', data=df)
plt.title("Online Delivery by Price Range")
plt.show()
```

```python
# 📙 LEVEL 2 - TASK 2: Price Range Analysis

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Dataset .csv")

# 🔢 Most Common Price Range
common_price = df['Price range'].value_counts()
print("Price Range Distribution:\n", common_price)

# 📊 Avg Rating per Price Range
avg_rating = df.groupby('Price range')['Aggregate rating'].mean()
print("\nAverage Ratings per Price Range:\n", avg_rating)

# 🎨 Colors and Ratings
color_rating = df.groupby('Rating color')['Aggregate rating'].mean().sort_values(ascending=False)
print("\nRating Color and Avg Ratings:\n", color_rating)

# 📊 Visualizations
sns.barplot(x=avg_rating.index, y=avg_rating.values)
plt.title("Average Rating per Price Range")
plt.xlabel("Price Range")
plt.ylabel("Avg Rating")
plt.show()

sns.countplot(x='Rating color', data=df, order=color_rating.index)
plt.title("Rating Colors Sorted by Avg Rating")
plt.show()
```
