```python
# üìò LEVEL 1 - TASK 1: Data Exploration and Preprocessing

# üì• Load Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# üìÇ Load Dataset
df = pd.read_csv("C:\\Users\\admin\\Downloads\\Dataset .csv")

# üîé Dataset Overview
print("Dataset Shape:", df.shape)
print("\nColumn Names:\n", df.columns)
print("\nData Types:\n", df.dtypes)

# üìâ Check for Missing Values
print("\nMissing Values:\n", df.isnull().sum())

# üõ†Ô∏è Handling Missing Values (Example: fill NA or drop)
# You can choose to drop or fill missing values based on column
df_cleaned = df.dropna()  # Or use df.fillna(method='ffill') or df['col'].fillna(value)

# üîÅ Data Type Conversion (if needed)
# Example: df['Aggregate rating'] = df['Aggregate rating'].astype(float)

# üéØ Analyze Target Variable: Aggregate Rating
print("\nAggregate Rating Value Counts:\n", df_cleaned['Aggregate rating'].value_counts())

# üìä Visualize Distribution of Target
plt.figure(figsize=(8, 5))
sns.countplot(x='Aggregate rating', data=df_cleaned, palette='viridis')
plt.title("Distribution of Aggregate Rating")
plt.xlabel("Aggregate Rating")
plt.ylabel("Count")
plt.show()

# üìå Summary
print("\nTarget Class Imbalance:")
print(df_cleaned['Aggregate rating'].value_counts(normalize=True) * 100)
```

```python
# üìò LEVEL 1 - TASK 2: Descriptive Analysis

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# üì• Load Data
df = pd.read_csv("C:\\Users\\admin\\Downloads\\Dataset .csv")

# üßÆ Basic Stats for Numerical Columns
print(df.describe())

# üó∫Ô∏è Explore Categorical Variables
print("\nUnique Countries:", df['Country Code'].nunique())
print("Unique Cities:", df['City'].nunique())
print("Unique Cuisines:", df['Cuisines'].nunique())

# üîù Top Cities by Restaurant Count
top_cities = df['City'].value_counts().head(10)
print("\nTop Cities:\n", top_cities)

# üîù Top Cuisines by Frequency
top_cuisines = df['Cuisines'].value_counts().head(10)
print("\nTop Cuisines:\n", top_cuisines)

# üìä Visualizations
plt.figure(figsize=(12, 5))
sns.countplot(y='City', data=df, order=df['City'].value_counts().iloc[:10].index, palette='coolwarm')
plt.title("Top 10 Cities with Most Restaurants")
plt.show()

plt.figure(figsize=(12, 5))
top_cuisines.plot(kind='barh', color='skyblue')
plt.title("Top 10 Cuisines")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisine Type")
plt.show()
```

```python
# üìò LEVEL 1 - TASK 3: Geospatial Analysis

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# üì• Load Data
df = pd.read_csv("C:\\Users\\admin\\Downloads\\Dataset .csv")

# üìç Check Latitude and Longitude
print("\nLatitude & Longitude Sample:")
print(df[['Latitude', 'Longitude']].head())

# üó∫Ô∏è Scatter Plot of Restaurant Locations
plt.figure(figsize=(10, 6))
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.4, c='red', s=10)
plt.title("Restaurant Locations")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()

# üß™ Location vs Rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Longitude', y='Latitude', hue='Aggregate rating', data=df, palette='viridis')
plt.title("Location vs Rating")
plt.show()

# üåç City-wise Distribution
plt.figure(figsize=(12, 6))
top_cities = df['City'].value_counts().head(10)
sns.barplot(x=top_cities.index, y=top_cities.values)
plt.title("Top 10 Cities by Restaurant Count")
plt.xticks(rotation=45)
plt.show()
```
